```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('imdb_top_1000.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies = movies[['Poster_Link', 'Series_Title', 'Released_Year', 'Runtime', 'Genre', 'IMDB_Rating', 'Overview', 'Director', 'Star1', 'Star2', 'Star3', 'Star4']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
type(movies.iloc[1].Overview)
```

```python
def convert(obj):
    L = []
    for i in obj.split(","):
        L.append(i)
    return L
```

```python
movies['Genre'] = movies['Genre'].apply(convert)
```

```python
movies['Genre']
```

```python
movies['Star1'] = movies['Star1'].apply(convert)
movies['Star2'] = movies['Star2'].apply(convert)
movies['Star3'] = movies['Star3'].apply(convert)
movies['Star4'] = movies['Star4'].apply(convert)
```

```python
movies['Cast'] = movies['Star1']+movies['Star2']+movies['Star3']+movies['Star4']
```

```python
movies['Director'] = movies['Director'].apply(convert)
```

```python
movies.head()
```

```python
movies['Overview'] = movies['Overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['Tags'] = movies['Overview']+movies['Genre']+movies['Cast']+movies['Director']
```

```python
movies.head()
```

```python
new_df = movies[['Poster_Link', 'Series_Title', 'Released_Year', 'Runtime', 'Tags']]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['Tags'][0]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similiraty = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['Series_Title']==movie].index[0]
    distances = similiraty[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].Series_Title, new_df.iloc[i[0]].Poster_Link)
```

```python
recommend('The Dark Knight')
```
